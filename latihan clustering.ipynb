```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Memuat dataset iris
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Clustering menggunakan KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(data)

# Clustering menggunakan DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(data)
```

```python
# Visualisasi hasil KMeans
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)', hue='KMeans_Cluster', palette='viridis', marker='o')
plt.title('KMeans Clustering')
plt.xlabel('Panjang Sepal (cm)')
plt.ylabel('Lebar Sepal (cm)')

# Visualisasi hasil DBSCAN
plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)', hue='DBSCAN_Cluster', palette='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('Panjang Sepal (cm)')
plt.ylabel('Lebar Sepal (cm)')

plt.tight_layout()
plt.show()
```

```python
# Hitung dan cetak silhouette score untuk masing-masing metode
kmeans_silhouette = silhouette_score(data.iloc[:, :-2], data['KMeans_Cluster'])
dbscan_silhouette = silhouette_score(data.iloc[:, :-2], data['DBSCAN_Cluster'])

print(f'Silhouette Score KMeans: {kmeans_silhouette}')
print(f'Silhouette Score DBSCAN: {dbscan_silhouette}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Buat dataset buatan
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Konversi data menjadi DataFrame
data = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])

# Clustering menggunakan KMeans
kmeans = KMeans(n_clusters=4)
data['KMeans_Cluster'] = kmeans.fit_predict(X)

# Clustering menggunakan DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(X)

# Visualisasi hasil KMeans
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(data['Feature 1'], data['Feature 2'], c=data['KMeans_Cluster'], cmap='viridis', marker='o')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Visualisasi hasil DBSCAN
plt.subplot(1, 2, 2)
plt.scatter(data['Feature 1'], data['Feature 2'], c=data['DBSCAN_Cluster'], cmap='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

# Hitung dan cetak silhouette score untuk masing-masing metode
kmeans_silhouette = silhouette_score(X, data['KMeans_Cluster'])
dbscan_silhouette = silhouette_score(X, data['DBSCAN_Cluster'])

print(f'Silhouette Score KMeans: {kmeans_silhouette}')
print(f'Silhouette Score DBSCAN: {dbscan_silhouette}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Simulate a dataset for 'Annual Income (k$)' and 'Spending Score (1-100)'
np.random.seed(0)
annual_income = np.random.randint(15, 137, size=300)
spending_score = np.random.randint(1, 101, size=300)

# Create a DataFrame with these features
data = pd.DataFrame({
    'Annual Income (k$)': annual_income,
    'Spending Score (1-100)': spending_score
})
# Visualisasi hasil KMeans
plt.figure(figsize=(12, 6))
```

```python
# Clustering using KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
data['KMeans_Cluster'] = kmeans.fit_predict(data[['Annual Income (k$)', 'Spending Score (1-100)']])

# Clustering using DBSCAN
dbscan = DBSCAN(eps=10, min_samples=5)
data['DBSCAN_Cluster'] = dbscan.fit_predict(data[['Annual Income (k$)', 'Spending Score (1-100)']])

# Visualizing KMeans Clustering
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], 
            c=data['KMeans_Cluster'], cmap='viridis', marker='o')
plt.title('KMeans Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Visualizing DBSCAN Clustering
plt.subplot(1, 2, 2)
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], 
            c=data['DBSCAN_Cluster'], cmap='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

plt.tight_layout()
plt.show()

# Calculate and print silhouette scores for both methods
kmeans_silhouette = silhouette_score(data[['Annual Income (k$)', 'Spending Score (1-100)']], 
                                     data['KMeans_Cluster'])

# Handle DBSCAN silhouette score calculation (ignore noise points with label -1)
dbscan_labels = data['DBSCAN_Cluster']
dbscan_valid = dbscan_labels != -1  # Exclude noise points

if len(set(dbscan_labels[dbscan_valid])) > 1:  # Ensure at least 2 clusters after filtering
    dbscan_silhouette = silhouette_score(
        data[['Annual Income (k$)', 'Spending Score (1-100)']][dbscan_valid],
        dbscan_labels[dbscan_valid]
    )
    print(f'Silhouette Score DBSCAN: {dbscan_silhouette}')
else:
    print('Silhouette Score DBSCAN: Not applicable (too few clusters after filtering noise points)')

print(f'Silhouette Score KMeans: {kmeans_silhouette}')
```
